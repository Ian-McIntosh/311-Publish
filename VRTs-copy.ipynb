Read in Hucs

```python
import fiona
import geopandas as gpd

# Get all the layers from the .gdb file 
layers = fiona.listlayers('wbdhu12_a_us_september2020-copy.gdb')
layers

for layer in layers:
    hucs = gpd.read_file('wbdhu12_a_us_september2020-copy.gdb',layer=0)
    # Do stuff with the gdf
hucs
```

Convert Hucs to EPSG:5070

```python
hucs.crs
hucs = hucs.to_crs("EPSG:5070")
hucs
```

Join hucs with city of Austin

```python
sjoined_listings = gpd.sjoin(hucs, city)
sjoined_listings
```

Get only unique hucs

```python
hucs = sjoined_listings['huc12'].tolist()
def unique(list1):
 
    # initialize a null list
    unique_list = []
     
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    return(unique_list)
allHucs = (unique(hucs))
```
